```python
#Load the dataset
df = pd.read_csv(Population, skiprows=4)
df1=df.head(30)
```

```python
# Display the first few rows of the dataframe to verify its structure
print(df1.head())
```

```python
# dislay the columns of the dataset
print(df1.columns)
```

```python
# Filter the data for the year 2020
population_2020 = df1[['Country Name', '2020']].dropna()
```

```python
# Rename columns
population_2020 = population_2020.rename(columns={'Country Name': 'Country', '2020': 'Population'})
```

```python
# Plotting
plt.figure(figsize=(12, 6))
plt.bar(population_2020['Country'], population_2020['Population'])
plt.title('Population Distribution by Country in 2020')
plt.xlabel('Country')
plt.ylabel('Population')
plt.xticks(rotation=90)
plt.grid(True)
plt.show()
```
